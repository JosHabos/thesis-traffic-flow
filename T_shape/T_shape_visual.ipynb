## Additional visualization for T-shape
The following notebook generates the Density Evolution Plot for the **T-Shaped Left and Right turn**.

#### Import Statements

```python
import os
import re
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

#### Visualization Font Adjustment

```python
plt.rcParams.update({
    "font.size": 14,
    "axes.titlesize": 16,
    "axes.labelsize": 16,
    "legend.fontsize": 10,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10
})
```

#### Initializing Parameters and Path

```python
# ==============================
# Simulation Parameters
# ==============================
T = 3.0
nSteps = 300

# ==============================
# Traffic Visualization Parameters
# ==============================
xmin = -20.0
xmax = 20.0
ymin = 20.0
ymax = 3.0

# Generates array of color values
colors = plt.cm.turbo(np.linspace(0, 1, 30))

# slice_length = 3.0
# cars_per_square = 5
# car_block_length = 0.4
# car_block_height = 0.2

# ==============================
# Modulo Selection
# ==============================
plot_modulo = 10

# ==============================
# Data Path
# ==============================
data_path = "T_shape_output"
```

#### Collection of Files
The following obtains the '.dat' files that contain density values at points (x,y).

```python
pattern = re.compile(r"density_k_(\d+)\.dat")

files = []
for filename in os.listdir(data_path):
    match = pattern.match(filename)
    if match:
        k = int(match.group(1))
        files.append((k, filename))

files.sort(key=lambda x: x[0])

if not files:
    raise RuntimeError("No density_k_*.dat files found in the directory.")
```

#### Density Evolution Plots
Generates a graph showing the evolution of traffic density on the road at specific timestamps, decided by `plot_modulo`.

##### Taking x-position:

Because the road is mainly vertical, taking the evolution plot over the x-axis is not necessary.

##### Taking y-position:

###### S_1 slice: veering to left road

```python
fig, ax = plt.subplots(figsize=(9, 6))

for k, filename in files:
    if k % plot_modulo != 0:
        continue

    t_k = (k / nSteps) * T

    full_path = os.path.join(data_path, filename)
    data = np.loadtxt(full_path)

    y_1 = data[:, 1]
    rho_1 = data[:, 2]

    i = int(k/10)

    ax.plot(y_1, rho_1, label=f"t={t_k:.3f}", color=colors[i])

ax.set_xlabel("Position y")
ax.set_xlim(ymax,ymin)
ax.invert_xaxis()
ax.set_ylabel("Density")
ax.grid(True)

output_file = os.path.join(
    data_path,
    f"S_1.density_evolution_yaxis_mod_{plot_modulo}.png"
)
plt.savefig(output_file, dpi=300, bbox_inches="tight")
plt.close()

print(f"Density evolution plot saved to: {output_file}")
```

###### S_2 slice: veering to the right road

```python
fig, ax = plt.subplots(figsize=(9, 6))

for k, filename in files:
    if k % plot_modulo != 0:
        continue

    t_k = (k / nSteps) * T

    full_path = os.path.join(data_path, filename)
    data = np.loadtxt(full_path)

    y_2 = data[:, 4]
    rho_2 = data[:, 5]

    i = int(k/10)

    ax.plot(y_2, rho_2, label=f"t={t_k:.3f}", color=colors[i])

ax.set_xlabel("Position y")
ax.set_xlim(ymax,ymin)
ax.invert_xaxis()
ax.set_ylabel("Density")
ax.grid(True)
ax.legend(loc="center left", bbox_to_anchor=(1.02, 0.5), ncol=2, fontsize='small')

output_file = os.path.join(
    data_path,
    f"S_2.density_evolution_yaxis_mod_{plot_modulo}.png"
)
plt.savefig(output_file, dpi=300, bbox_inches="tight")
plt.close()

print(f"Density evolution plot saved to: {output_file}")
```
